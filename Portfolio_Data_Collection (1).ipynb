```python
all_data = [soxx_data, pick_data, mxi_data, futy_data, pave_data, tan_data,
            dtcr_data, volt_data]

combined_data = pd.concat(all_data, axis=1)
combined_data.head()
print(combined_data.columns)
```

```python
combined_data.shape
```

```python
combined_data.columns.duplicated()
combined_data = combined_data.loc[:, ~combined_data.columns.duplicated()]   #https://stackoverflow.com/questions/14984119/python-pandas-remove-duplicate-columns
print(combined_data.columns.duplicated())
print(combined_data.columns)
combined_data.head(10)
```

```python
fx_map = {
    '.SR': 'SARUSD=X', '.L': 'GBPUSD=X', '.F': 'EURUSD=X',
    '.BE': 'EURUSD=X', '.MX': 'MXNUSD=X', '.MU': 'EURUSD=X',
    '.TW': 'TWDUSD=X', '.TA': 'ILSUSD=X', '.MC': 'EURUSD=X',
    '.KS': 'KRWUSD=X', '.HK': 'HKDUSD=X', '.SI': 'SGDUSD=X'
}

req_fx = list(set(fx_map.values()))
fx_data = yf.download(req_fx, start=start_date, end=end_date)['Close']

port_data_usd = pd.DataFrame(index=combined_data.index)

for ticker in combined_data.columns:
    suffix = next((s for s in fx_map.keys() if ticker.endswith(s)), None)

    if suffix:
        fx_ticker = fx_map[suffix]

        fx_rate = fx_data[fx_ticker].reindex(combined_data.index).ffill()

        if suffix == ['.L', '.TA']:
            port_data_usd[ticker] = (combined_data[ticker] / 100) * fx_rate       #.L Prices often in Pence
        else:                                                                     #1 Israeli New Shekel = 100 Agora -> .TA
            port_data_usd[ticker] = combined_data[ticker] * fx_rate

        print(f"Converted {ticker} ({suffix}) to USD")
    else:
        port_data_usd[ticker] = combined_data[ticker]
```

```python
import numpy as np

#soxx_ret = (soxx_data.pct_change()) * 100     #Test
#soxx_vol = np.sqrt(252) * soxx_ret.std()
#soxx_vol.head(50)

combined_ret = (combined_data.pct_change()) * 100

combined_vol = np.sqrt(252) * combined_ret.std()    #annualised volatility

combined_vol.head(20)
```

```python
!pip install arch
from arch import arch_model

garch_results = {}

for col in combined_ret.columns:
    series = combined_ret[col].dropna()

    if series.empty:
        print(f"Skipping {col} because it is empty.")
        continue

    model = arch_model(
        series,           #arch_model requires 1D input
        p=1,              #p is the number of lagged/delayed volatility terms. Represents how long it takes shock-induced volatility to decay (1=yesterday's forecast)
        q=1,              #q is the number of lagged residuals/errors. Represents how impactful the news is on volatility (1=yesterday's news)
        mean='constant',  #Mean does not change over time (Assumption)
        vol='GARCH',      #GARCH opposed to ARCH (G=Generalised) adds the p term, and therefore generalises volatility forecast over time
        dist='normal'     #Assume returns are normally distributed
    )                     #GARCH(p=2, q=2) or over often overfits

    res = model.fit(disp="off")
    garch_results[col] = res
    forecast = res.forecast(horizon=5)
    print(col)
    print(forecast.variance.iloc[-1])
    print("-" * 40)
```

```python
trad_week_vols = {}

for col, res in garch_results.items():
    forecast = res.forecast(horizon=5)
    trading_week_vol = np.sqrt(forecast.variance.iloc[-1].sum())
    trad_week_vols[col] = trading_week_vol

trad_week_vol_data = (pd.Series(trad_week_vols, name="trad_week_vol").sort_values(ascending=False))
```
